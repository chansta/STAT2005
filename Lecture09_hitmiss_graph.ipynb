# Hit and Miss Graphs for Lecture 09

This notebook contains the code generating the graphs to explain hit and miss. 

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = [12,9]
```

```python
xset = np.arange(0,1.25,0.01)
yset = np.arange(0,1,0.01)
xbound = np.arange(0,1.01,0.01)
T = len(xset)
Ty = len(yset)
yf = lambda x: 1-np.power(x,2)
y = yf(xset)
fig,ax = plt.subplots()
ax.set_xlim(0,1.3)
ax.set_ylim(-0.7,1.1)
ax.plot(xset,y, color='r')
ax.plot(xbound, np.zeros(len(xbound)), color='b')
ax.plot(xbound, np.ones(len(xbound)), color='b')
ax.plot(np.ones(Ty), yset, color='b')
ax.fill_between(xset[0:100], y[0:100], color=(1,0.8,1))
ax.set_title('Integrating betwee 0 and 1 with $c=0$')
plt.savefig('./pics/c0.png')
```

```python
lx=0.8
xset = np.arange(0,1.25,0.01)
yset = np.arange(0,1,0.01)
xbound = np.arange(0,lx+0.01,0.01)
T = len(xset)
Ty = len(yset)
yf = lambda x: 1-np.power(x,2)
y = yf(xset)
fig,ax = plt.subplots()
ax.set_xlim(0,1.3)
ax.set_ylim(-0.7,1.1)
ax.plot(xset,y, color='r')
ax.plot(xbound, yf(lx)*np.ones(len(xbound)), color='b')
ax.plot(xset, np.zeros(T), color=(0,0,0))
ax.plot(xbound, np.ones(len(xbound)), color='b')
ax.plot(lx*np.ones(Ty), yset, color='b')
ax.fill_between(xbound, yf(xbound), yf(lx)*np.ones(len(xbound)), color=(1,0.8,1))
ax.fill_between(xbound, np.zeros(len(xbound)), yf(lx)*np.ones(len(xbound)), color=(0,0.8,0))
ax.set_title('Integrating betwee 0 and 0.8 with $c={0:.4}$'.format(yf(lx)))
ax.annotate(xy=(0.81,yf(lx)), s='$c={0:.4}$'.format(yf(lx)))
ax.annotate(xy=(0.4,0.2), s='$c(b-a)>0$')
plt.savefig('./pics/cgeq0.png')
```

```python
lx=1.2
xset = np.arange(0,1.25,0.01)
xbound = np.arange(0,lx+0.01,0.01)
T = len(xset)
yf = lambda x: 1-np.power(x,2)
y = yf(xset)
Ty = len(yset)
yset = np.arange(yf(lx),1,0.01)
fig,ax = plt.subplots()
ax.set_xlim(0,1.3)
ax.set_ylim(-0.7,1.1)
ax.plot(xset,y, color='r')
ax.plot(xbound, yf(lx)*np.ones(len(xbound)), color='b')
ax.plot(xset, np.zeros(T), color=(0,0,0))
ax.plot(xbound, np.ones(len(xbound)), color='b')
ax.plot(lx*np.ones(len(yset)), yset, color='b')
ax.fill_between(xbound, yf(xbound), yf(lx)*np.ones(len(xbound)), color=(1,0.8,1))
ax.fill_between(xbound, np.zeros(len(xbound)), yf(lx)*np.ones(len(xbound)), color=(0,0.8,0))
ax.fill_between(np.arange(1,1.21,0.01), yf(np.arange(1,1.21,0.01)), color=(0,0.8,1))
ax.set_title('Integrating betwee 0 and 1.2 with $c={0:.4}$'.format(yf(lx)))
ax.annotate(xy=(0.8,yf(lx)-0.1), s='$c={0:.4}$'.format(yf(lx)))
ax.annotate(xy=(0.4,-0.2), s='$c(b-a)<0$')
plt.savefig('./pics/cleq0.png')
```
